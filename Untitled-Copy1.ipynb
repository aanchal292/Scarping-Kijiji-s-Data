```python
# import the necessary libraries
import requests
import time
import pandas as pd 
from bs4 import BeautifulSoup
```

## Scraping data on rental houses and rooms from Kijiji and storing it in a dataframe

```python
price=[]
title=[]
utilities=[]
location=[]
description=[]
distance=[]
date=[]
```

```python
pages_to_get=4000
```

```python
for page in range(1, pages_to_get):
    print('processing page:', page)

    url = 'https://www.kijiji.ca/b-for-rent/city-of-toronto/page-'+str(page)+'/c30349001l1700273?radius=200.0&address=Toronto%2C+ON&ll=43.653226,-79.383184'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')

    ads = soup.find_all("div", attrs={'class': "search-item  top-feature "})

    # location of title
    title_source = soup.find_all("div", attrs={'class': "title"})
    location_source = soup.find_all("div", attrs={'class': 'location'})
    information_source = soup.find_all("div", attrs={'class':'description'})
    price_source = soup.find_all("div", attrs={'class':'price'})
    distance_source= soup.find_all("div",attrs={'class':'distance'})
    date_source = soup.find_all("span", attrs={'class':'date-posted'})

    # Getting all the titles
    for i in title_source:
        title_source_link = i.find_all('a')
        if title_source_link:
            statement_link = title_source_link[0].text.strip()
            title.append(statement_link)

    # Getting all the locations
    for l in location_source:
        location_source_link = l.find_all('span')
        if location_source_link:
            location_link = location_source_link[0].text.strip()
            location.append(location_link)
    
    #getting info
    
    for info in information_source:
        information_link = info.text.strip()
        description.append(information_link)
        
        
   # getting distance
    
    for dist in distance_source:
        distance_source_link = dist.text.strip()
        distance.append(distance_source_link)
        
   # date posted
   
    for dat in date_source:
        date_source_link = dat.text.strip()
        date.append(date_source_link)
        
   


    # getting price
    
    for p in price_source:
        price_link = p.text.strip()
        price.append(price_link)

        
```

```python
# Adjusting lengths of title, price, and utilities lists to match
min_length = min(len(title), len(location), len(description), len(price),len(distance), len(date))
title = title[:min_length]
location = location[:min_length]
distance=distance[:min_length]
description = description[:min_length]
date=date[:min_length]
price = price [:min_length]


```

```python
# creating dataframe
df = pd.DataFrame(columns=['Title','Location','Distance','Description','Date', 'Price'])
df['Title']=title

df['Location'] = location

df['Distance'] = distance

df['Description'] =description

df['Date']=date

df['Price'] = price
```

### Dataset is scraped and consists of the following columns 
1. Title
2. Location
3. Distance
4. Description
5. Date
6. Price


```python
# Save the DataFrame as a CSV file
df.to_csv('scraped_data.csv', index=False)
```
